```python
def fix_encoding(name: str) -> str:
    try:
        fixed_name = name.encode('ISO-8859-1').decode('UTF-8')
    except UnicodeError:
        # If encoding or decoding fails, return the original name
        fixed_name = name
    return fixed_name

# Open the file for reading
with open('bWAR.txt', encoding='utf-8') as f:
    # Create an empty dictionary to store the data
    bplayer_data = {}

    # Loop through each line in the file
    for line in f:
        # Split the line into fields using a comma as the delimiter
        fields = line.strip().split(",")

        # Extract the player name and WAR from the fields
        player_name = fields[1]
        war = float(fields[2])

        # Fix the encoding issue in the player name
        fixed_player_name = fix_encoding(player_name)

        # Add the player data to the dictionary
        bplayer_data[fixed_player_name] = war

# Print the dictionary
print(bplayer_data)

def lower(a, b):
    if a < b:
        return a
    else:
        return b
```

```python
# Open the file for reading
def fix_encoding(name: str) -> str:
    try:
        fixed_name = name.encode('ISO-8859-1').decode('UTF-8')
    except UnicodeError:
        # If encoding or decoding fails, return the original name
        fixed_name = name
    return fixed_name

# Print the dictionary
#print(fplayer_data)

with open("fWAR2.txt", "r") as f:
    # Create an empty dictionary to store the data
    fplayer_data2 = {}

    # Loop through each line in the file
    for line in f:
        # Split the line into fields using a tab as the delimiter
        fields = line.strip().split("\t")

        # Extract the player name and WAR from the fields
        player_name = fields[0].replace('"', '')
        war = float(fields[6].replace('"', ''))

        # Fix the encoding issue in the player name
        fixed_player_name = fix_encoding(player_name)

        # Add the player data to the dictionary
        fplayer_data2[fixed_player_name] = war
        
print(fplayer_data2)
```

```python
def add_dicts(dict1, dict2):
    result_dict = {}

    # Loop through each key in dict1
    for key in dict1:
        # Add the value for the key in dict1 to the result_dict
        result_dict[key] = dict1[key]

        # If the key is also in dict2, add the value from dict2 to the result_dict
        if key in dict2:
            result_dict[key] += dict2[key]

    # Loop through each key in dict2 that isn't in dict1
    for key in dict2:
        if key not in dict1:
            # Add the value for the key in dict2 to the result_dict
            result_dict[key] = dict2[key]

    return result_dict

warscore = add_dicts(bplayer_data,fplayer_data2)
print(warscore)
```

```python
def round_dict_values(d):
    """
    Rounds all values in a dictionary to 1 decimal place.

    Args:
        d (dict): A dictionary with numeric values.

    Returns:
        dict: A new dictionary with the same keys as the input dictionary, but with all values rounded to 1 decimal place.
    """
    return {k: round(v, 1) for k, v in d.items()}
morescore = round_dict_values(warscore)
```
