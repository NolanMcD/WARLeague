```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for LHB vs LHP (Left-Handed Batters vs Left-Handed Pitchers)
data_lhp = {
    'Player': ['Yordan Alvarez', 'Jeff McNeil', 'Steven Kwan', 'Luis Arraez', 'Freddie Freeman', 
               'Kyle Tucker', 'Corey Seager', 'Bryce Harper', 'Nathaniel Lowe', 'Juan Soto'],
    'BA': [0.315, 0.289, 0.285, 0.274, 0.273, 0.271, 0.270, 0.269, 0.269, 0.268],
    'PA': [957, 904, 586, 773, 2735, 980, 1448, 2271, 898, 1372]
}

# Data for RHB vs RHP (Right-Handed Batters vs Right-Handed Pitchers)
data_rhp = {
    'Player': ['Mike Trout', 'Jose Altuve', 'Aaron Judge', 'Trea Turner', 'Ronald Acuna Jr.',
               'Mookie Betts', 'Vladimir Guerrero Jr.', 'Bobby Witt Jr.', 'Bo Bichette', 'Justin Turner'],
    'BA': [0.301, 0.297, 0.296, 0.294, 0.293, 0.292, 0.292, 0.289, 0.288, 0.288],
    'PA': [5033, 6104, 3443, 3852, 2541, 4850, 2969, 1800, 2326, 4403]
}

# Convert data to DataFrame
df_lhp = pd.DataFrame(data_lhp)
df_rhp = pd.DataFrame(data_rhp)

# Create scatter plot for LHP vs LHB
plt.figure(figsize=(10, 6))

# Plot LHP vs LHB
plt.scatter(df_lhp['PA'], df_lhp['BA'], color='blue', label='LHB vs LHP', s=100)
plt.scatter(df_rhp['PA'], df_rhp['BA'], color='red', label='RHB vs RHP', s=100)

# Adding labels and title
plt.title('Batting Average vs Plate Appearances for LHB vs LHP and RHB vs RHP')
plt.xlabel('Plate Appearances (PA)')
plt.ylabel('Batting Average (BA)')
plt.legend()

# Annotate players on the scatter plot
for i, player in enumerate(df_lhp['Player']):
    plt.annotate(player, (df_lhp['PA'][i], df_lhp['BA'][i]), fontsize=9, color='blue', alpha=0.7)

for i, player in enumerate(df_rhp['Player']):
    plt.annotate(player, (df_rhp['PA'][i], df_rhp['BA'][i]), fontsize=9, color='red', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the movie watching methods with accurate numbers
labels = ['Argh Matey (Piracy)', 'In Theatres (AMC)', 'In Theartes (Cosford)', 'HBOMax', 'TV On Demand', 'Netflix', 
          'Prime Video', 'Criterion', 'On a Plane', 'My Collection', 'Mubi', 'Peacock', 
          'Tubi', 'YouTube', 'TV', 'Ureserves', 'AppleTV', 'PlutoTV', 
          'Disney+', 'Hulu', 'Read the Screenplay']

sizes = [194, 184, 15, 104, 93, 85, 71, 67, 36, 28, 25, 20, 19, 19, 17, 16, 14, 13, 9, 9, 6]

# Colors for each section with adjustments based on your requests
colors_adjusted = ['#000000',  # Argh Matey (Piracy) - black
                   '#c73010',  # In Theatres - red
                   '#d2320f',  # Cosford - darker red
                   '#187cd5',  # HBOMax - light blue
                   '#3385FF',  # TV On Demand - yellow
                   '#3366FF',  # Netflix - dark blue
                   '#335CFF',  # Prime Video - navy blue
                   '#334FFF',  # Criterion - ocean blue
                   '#3339FF',  # On a Plane - light green
                   '#3326FF',  # My Collection - light green
                   '#3313FF',  # Mubi - light green
                   '#32CD32',  # Peacock - light green
                   '#FFEB3B',  # Tubi - light yellow
                   '#FF9800',  # YouTube - light green
                   '#9E9E9E',  # TV - light yellow
                   '#FF9933',  # Ureserves - light green
                   '#6699FF',  # AppleTV - light blue
                   '#FF6666',  # PlutoTV - light green
                   '#00CCFF',  # Disney+ - light blue
                   '#00FF00',  # Hulu - light green
                   '#6A1B9A']  # Read the Screenplay - purple

# Plotting the pie chart with gray background and white text
plt.figure(figsize=(13, 13), facecolor='gray')

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors_adjusted, textprops={'fontsize': 12, 'color': 'white'})

# Title and background
plt.title('How I Watch My Movies (Accurate Breakdown)', fontsize=16, color='white')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is a circle.

# Adjusting the position of the legend (key) to top-left without overlap
plt.legend(['Piracy', 'In Theatres', 'Paid Streaming', 'Free Streaming', 'Cable TV'], 
           loc='upper left', 
           fontsize=12,  
           framealpha=0.5, 
           title='Legend', 
           title_fontsize='13', 
           facecolor='white',  # Adjust legend background color here
           bbox_to_anchor=(0.9999, 0.99999))

# Show the plot
plt.show()
```
