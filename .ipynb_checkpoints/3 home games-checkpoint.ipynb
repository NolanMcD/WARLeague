```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Read the CSV file without header
df = pd.read_csv('3homerdata.txt', header=None, names=["Rk", "Player", "HR", "Date", "Player-additional"])

# Clean the Date column by removing any text after the date (e.g., " (2)")
df['Date'] = df['Date'].apply(lambda x: re.sub(r'\s*\(.*\)', '', x))

# Convert the Date column to datetime format
df['Year'] = pd.to_datetime(df['Date'], errors='coerce').dt.year

# Drop any rows where the date couldn't be parsed
df = df.dropna(subset=['Year'])

# Create a column to differentiate between 3 HR and 4 HR games
df['HR_Type'] = df['HR'].apply(lambda x: '3 HR' if x == 3 else '4 HR')

# Group by Year and HR_Type to count the number of occurrences
yearly_counts = df.groupby(['Year', 'HR_Type']).size().unstack(fill_value=0)

# Plotting with a larger image size
fig, ax = plt.subplots(figsize=(15, 10))

# Colors for 3 HR and 4 HR games
colors = ['#1f77b4', '#ff7f0e']

# Stacked bar chart by year
yearly_counts.plot(kind='bar', stacked=True, color=colors, ax=ax)

# Adding labels and title
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Number of 3+ Home Run Games', fontsize=14)
ax.set_title('3+ Home Run Games by Year in MLB History', fontsize=16)

# Improve legend
ax.legend(title='Home Run Type', title_fontsize='13', fontsize='11')

# Enhance tick labels for better readability
ax.tick_params(axis='x', labelrotation=90)
ax.tick_params(axis='y', labelsize=12)
ax.tick_params(axis='x', labelsize=12)

# Add gridlines for better readability
ax.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7)

# Adjust margins and spacing to make the plot look better
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

# Display the plot
plt.show()
```
